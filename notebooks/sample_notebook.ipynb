# Sample Notebook

This notebook demonstrates LaTeX and plots.

Equation: $y = \sin(x)$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)")
plt.title("Sample Plot")
plt.legend()
plt.savefig("../assets/images/sample_plot.png")
plt.show()
```
